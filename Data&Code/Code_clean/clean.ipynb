```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/Arabic_Toxicity_Dataset.csv")
```

```python
df
```

```python

# Extraire les colonnes "Text" et "Toxic/Not Toxic"
df_clean = df[['Text', 'Toxic/Not Toxic']]
df_clean

```

```python
# Remplacer les valeurs "Toxic" par 1 et "Not Toxic" par 0
df_clean['Toxic/Not Toxic'] = df_clean['Toxic/Not Toxic'].apply(lambda x: 1 if x == 'Toxic' else 0)
```
