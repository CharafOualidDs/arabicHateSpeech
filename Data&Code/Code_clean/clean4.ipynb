```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("data/TweetClassification-Summary.xlsx")
```

```python
df
```

```python

# Extraire les colonnes "Text" et "Toxic/Not Toxic"
df1 = df[['text', 'aggregatedAnnotation']]
df1

```

```python

df_clean = df1.rename(columns={'aggregatedAnnotation': 'Toxic/Not Toxic' , 'text': 'Text'})

df_clean
```

```python
# Remplacer les valeurs "Toxic" par 1 et "Not Toxic" par 0
df_clean['Toxic/Not Toxic'] = df_clean['Toxic/Not Toxic'].apply(lambda x: 0 if x == 0 else 1)
```

```python
df_clean
```

```python
# Get the unique values in 'column1'
unique_values = df_clean['Toxic/Not Toxic'].unique()

# Print the unique values
print("Unique values in column1:", unique_values)
```

```python
num_toxic = df_clean['Toxic/Not Toxic'].value_counts()[1]
print(f"Il y a {num_toxic} lignes étiquetées comme toxiques.")
num_toxic = df_clean['Toxic/Not Toxic'].value_counts()[0]
print(f"Il y a {num_toxic} lignes étiquetées comme non toxiques.")
```

```python
# Enregistrer la dataframe nettoyée dans un fichier CSV
df_clean.to_csv('clean/TweetClassification-Summary_nettoye.csv', index=False)
```
