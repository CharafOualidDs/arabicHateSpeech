```python
import os
import pandas as pd
```

```python
# Chemin du dossier contenant les fichiers CSV
folder_path = 'clean'

# Créer une liste pour stocker les dataframes
dfs = []
```

```python
# Boucle sur les fichiers CSV dans le dossier et ajouter chaque dataframe à la liste
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        filepath = os.path.join(folder_path, filename)
        df = pd.read_csv(filepath)
        dfs.append(df)
```

```python
# Concaténer toutes les dataframes dans la liste en une seule dataframe
merged_df = pd.concat(dfs, ignore_index=True)

# Enregistrer la dataframe fusionnée dans un nouveau fichier CSV
merged_df.to_csv('clean_fusionne.csv', index=False)
```
