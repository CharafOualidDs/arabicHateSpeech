```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("dev_data.csv")
```

```python
df.head(10)
```

```python
def is_toxic(row):
    if row['Offensive'] == 'OFF' or row['Hate'] == 'HS':
        return 1
    else:
        return 0
```

```python
df['Toxic/Not Toxic'] = df.apply(is_toxic, axis=1)
```

```python

# Extraire les colonnes "Text" et "Toxic/Not Toxic"
df_clean = df[['Tweet', 'Toxic/Not Toxic']]
#df_clean
df_clean = df_clean.rename(columns={'Tweet': 'Text' })
df_clean


```
